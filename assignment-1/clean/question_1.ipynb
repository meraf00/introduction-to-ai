```python
import os

import matplotlib.pyplot as plt
%matplotlib inline

from graph import *
import utils
```

### Load map

```python
romania_coords_filepath = os.path.join(os.getcwd(), "romania-data/romania-coords.txt")
romania_map_filepath = os.path.join(os.getcwd(), "romania-data/romania-road-distance.txt")

romania_map = utils.load_city_graph(romania_map_filepath)
romania_coords = utils.load_coords(romania_coords_filepath)
```

```python
for city in romania_map:
    
    y1, x1 = romania_coords[city]
    
    plt.text(x1 + 0.1, y1 + 0.1, city)

    for neighbour, _ in romania_map.neighbours(city):
        
        y2, x2 = romania_coords[neighbour]

        plt.plot([x1, x2], [y1, y2], marker="o")
```
