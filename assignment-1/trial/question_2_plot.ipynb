```python
from matplotlib import pyplot as plt
import pandas as pd

from collections import defaultdict
from Romania import road_map, h_longitude_latitude_distance

df = pd.read_csv('result.csv')
df.head(60)
df[(df["Algorithm"] == 'uniform_cost_search') | (df["Algorithm"] == 'a_star_search')].head(30)
```

```python
sample_cities = []

#            bfs dfs ucs  id  gr  a*
algorithms = [[], [], [], [], [], []]


# averg
cities_heuristic = []

for i in range(0, len(df), 6):


    cities = (df.iloc[i][1], df.iloc[i][2])
    sample_cities.append(cities)

    actual_distance = h_longitude_latitude_distance(cities[0], cities[1])
    cities_heuristic.append(actual_distance)
    

    for j in range(6):
        algorithms[j].append(df.iloc[i + j][3])




algs = ['breadth_first_search', 'depth_first_search', 'uniform_cost_search', 'iterative_deepening', 'greedy_search', 'a_star_search']

for i in range(6):
    plt.plot(cities_heuristic, algorithms[i], label=algs[i])

plt.show
```
